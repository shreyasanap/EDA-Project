Lets first know what is skewness ? 🥴
- Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean

Skewness can be right(positive),left(negative),zero🤯

1. Right Skewness 👉 : A few extremely high values can pull the mean towards them, causing the distribution to be skewed to the right.

2. Left Skewness 👈: A few extremely low values can drag the mean towards them, causing the distribution to be skewed to the left.

### The student having highest 10th % theres a probability that it will have high 12th % & college GPA ⚡⚡

Chi Squared Statistic💡
Why it is performed?
1. It determines if there's a significant association between two categorical variables.
2. It helps in understanding if changes in one variable are related to changes in another.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
cont_table = pd.crosstab(df['Salary'], df['Designation'])

# Perform chi-squared test of independence
chi_stat, p_val, dof, expected = chi2_contingency(cont_table)

# Print results
print("Chi-squared statistic:", chi_stat)
print("p-value:", p_val)

# Set significance level
alpha = 0.05
print("\nSignificance level:", alpha)

# Interpret results
if p_val < alpha:
    print("Reject the null hypothesis: There is a relationship between salary and designation.")
else:
    print("Fail to reject the null hypothesis: There is no relationship between salary and designation.")
```

```python
import scipy.stats as stats
male_salary = df[df['Gender'] == 'Male']['Salary']
female_salary = df[df['Gender'] == 'Female']['Salary']
t_stat, p_val = stats.ttest_ind(male_salary, female_salary, equal_var=False)
if p_val < 0.05:
  print("There is a significant difference in salary between genders.")
else:
  print("There is no significant difference in salary between genders.")
```
